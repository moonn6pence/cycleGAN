```python
from PIL import Image
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import tensorflow_addons as tfa
import matplotlib.pyplot as plt
import numpy as np
import os

AUTOTUNE = tf.data.experimental.AUTOTUNE
```

```python
def _bytes_feature(value):
  if isinstance(value, type(tf.constant(0))):
    value = value.numpy() 
  return tf.train.Feature(bytes_list=tf.train.BytesList(value=[value]))
```

```python
def image_example(image_string):
  image_shape = tf.image.decode_jpeg(image_string).shape

  feature = {
      'image': _bytes_feature(image_string),
  }

  return tf.train.Example(features=tf.train.Features(feature=feature))
```

```python
IMAGE_SIZE = [256, 256]

def decode_image(image):
    image = tf.image.decode_jpeg(image, channels=3)
    image = (tf.cast(image, tf.float32) / 127.5) - 1
    image = tf.reshape(image, [*IMAGE_SIZE, 3])
    return image

def read_tfrecord(example):
    tfrecord_format = {
        "image": tf.io.FixedLenFeature([], tf.string),
    }
    example = tf.io.parse_single_example(example, tfrecord_format)
    image = decode_image(example['image'])
    return image
```

```python
def load_dataset(filenames, labeled=True, ordered=False):
    dataset = tf.data.TFRecordDataset(filenames)
    dataset = dataset.map(read_tfrecord, num_parallel_calls=AUTOTUNE)
    return dataset
```

```python
OUTPUT_CHANNELS = 3

def downsample(filters, size, apply_instancenorm=True):
    initializer = tf.random_normal_initializer(0., 0.02)
    gamma_init = keras.initializers.RandomNormal(mean=0.0, stddev=0.02)

    result = keras.Sequential()
    result.add(layers.Conv2D(filters, size, strides=2, padding='same',
                             kernel_initializer=initializer, use_bias=False))

    if apply_instancenorm:
        result.add(tfa.layers.InstanceNormalization(gamma_initializer=gamma_init))

    result.add(layers.LeakyReLU())

    return result
```

```python
def upsample(filters, size, apply_dropout=False):
    initializer = tf.random_normal_initializer(0., 0.02)
    gamma_init = keras.initializers.RandomNormal(mean=0.0, stddev=0.02)

    result = keras.Sequential()
    result.add(layers.Conv2DTranspose(filters, size, strides=2,
                                      padding='same',
                                      kernel_initializer=initializer,
                                      use_bias=False))

    result.add(tfa.layers.InstanceNormalization(gamma_initializer=gamma_init))

    if apply_dropout:
        result.add(layers.Dropout(0.5))

    result.add(layers.ReLU())

    return result
```

```python
def Generator():
    inputs = layers.Input(shape=[256,256,3])

    down_stack = [
        downsample(64, 4, apply_instancenorm=False),
        downsample(128, 4),
        downsample(256, 4),
        downsample(512, 4),
        downsample(512, 4),
        downsample(512, 4), 
        downsample(512, 4),
        downsample(512, 4),
    ]

    up_stack = [
        upsample(512, 4, apply_dropout=True),
        upsample(512, 4, apply_dropout=True),
        upsample(512, 4, apply_dropout=True),
        upsample(512, 4),
        upsample(256, 4),
        upsample(128, 4),
        upsample(64, 4),
    ]

    initializer = tf.random_normal_initializer(0., 0.02)
    last = layers.Conv2DTranspose(OUTPUT_CHANNELS, 4,
                                  strides=2,
                                  padding='same',
                                  kernel_initializer=initializer,
                                  activation='tanh') 

    x = inputs
    
    skips = []
    for down in down_stack:
        x = down(x)
        skips.append(x)

    skips = reversed(skips[:-1])

    for up, skip in zip(up_stack, skips):
        x = up(x)
        x = layers.Concatenate()([x, skip])

    x = last(x)

    return keras.Model(inputs=inputs, outputs=x)
```

```python
def Discriminator():
    initializer = tf.random_normal_initializer(0., 0.02)
    gamma_init = keras.initializers.RandomNormal(mean=0.0, stddev=0.02)

    inp = layers.Input(shape=[256, 256, 3], name='input_image')

    x = inp

    down1 = downsample(64, 4, False)(x)
    down2 = downsample(128, 4)(down1)
    down3 = downsample(256, 4)(down2)

    zero_pad1 = layers.ZeroPadding2D()(down3)
    conv = layers.Conv2D(512, 4, strides=1,
                         kernel_initializer=initializer,
                         use_bias=False)(zero_pad1)

    norm1 = tfa.layers.InstanceNormalization(gamma_initializer=gamma_init)(conv)

    leaky_relu = layers.LeakyReLU()(norm1)

    zero_pad2 = layers.ZeroPadding2D()(leaky_relu)

    last = layers.Conv2D(1, 4, strides=1,
                         kernel_initializer=initializer)(zero_pad2)

    return tf.keras.Model(inputs=inp, outputs=last)
```

```python
strategy = tf.distribute.get_strategy()
with strategy.scope():
    old_generator = Generator()
    young_generator = Generator()

    old_discriminator = Discriminator()
    young_discriminator = Discriminator()
```

```python
class CycleGan(keras.Model):
    def __init__(
        self,
        old_generator,
        young_generator,
        old_discriminator,
        young_discriminator,
        lambda_cycle=10,
    ):
        super(CycleGan, self).__init__()
        self.o_gen = old_generator
        self.y_gen = young_generator
        self.o_disc = old_discriminator
        self.y_disc = young_discriminator
        self.lambda_cycle = lambda_cycle
        
    def compile(
        self,
        o_gen_optimizer,
        y_gen_optimizer,
        o_disc_optimizer,
        y_disc_optimizer,
        gen_loss_fn,
        disc_loss_fn,
        cycle_loss_fn,
        identity_loss_fn
    ):
        super(CycleGan, self).compile()
        self.o_gen_optimizer = o_gen_optimizer
        self.y_gen_optimizer = y_gen_optimizer
        self.o_disc_optimizer = o_disc_optimizer
        self.y_disc_optimizer = y_disc_optimizer
        self.gen_loss_fn = gen_loss_fn
        self.disc_loss_fn = disc_loss_fn
        self.cycle_loss_fn = cycle_loss_fn
        self.identity_loss_fn = identity_loss_fn
        
    def train_step(self, batch_data):
        real_old, real_young = batch_data
        
        with tf.GradientTape(persistent=True) as tape:
            fake_old = self.o_gen(real_young, training=True)
            cycled_young = self.y_gen(fake_old, training=True)

            fake_young = self.y_gen(real_old, training=True)
            cycled_old = self.o_gen(fake_young, training=True)

            same_old = self.o_gen(real_old, training=True)
            same_young = self.y_gen(real_young, training=True)

            disc_real_old = self.o_disc(real_old, training=True)
            disc_real_young = self.y_disc(real_young, training=True)

            disc_fake_old = self.o_disc(fake_old, training=True)
            disc_fake_young = self.y_disc(fake_young, training=True)

            old_gen_loss = self.gen_loss_fn(disc_fake_old)
            young_gen_loss = self.gen_loss_fn(disc_fake_young)

            total_cycle_loss = self.cycle_loss_fn(real_old, cycled_old, self.lambda_cycle) + self.cycle_loss_fn(real_young, cycled_young, self.lambda_cycle)

            total_old_gen_loss = old_gen_loss + total_cycle_loss + self.identity_loss_fn(real_old, same_old, self.lambda_cycle)
            total_young_gen_loss = young_gen_loss + total_cycle_loss + self.identity_loss_fn(real_young, same_young, self.lambda_cycle)

            old_disc_loss = self.disc_loss_fn(disc_real_old, disc_fake_old)
            young_disc_loss = self.disc_loss_fn(disc_real_young, disc_fake_young)

        old_generator_gradients = tape.gradient(total_old_gen_loss,
                                                  self.o_gen.trainable_variables)
        young_generator_gradients = tape.gradient(total_young_gen_loss,
                                                  self.y_gen.trainable_variables)

        old_discriminator_gradients = tape.gradient(old_disc_loss,
                                                      self.o_disc.trainable_variables)
        young_discriminator_gradients = tape.gradient(young_disc_loss,
                                                      self.y_disc.trainable_variables)

        self.o_gen_optimizer.apply_gradients(zip(old_generator_gradients,
                                                 self.o_gen.trainable_variables))

        self.y_gen_optimizer.apply_gradients(zip(young_generator_gradients,
                                                 self.y_gen.trainable_variables))

        self.o_disc_optimizer.apply_gradients(zip(old_discriminator_gradients,
                                                  self.o_disc.trainable_variables))

        self.y_disc_optimizer.apply_gradients(zip(young_discriminator_gradients,
                                                  self.y_disc.trainable_variables))
        
        return {
            "old_gen_loss": total_old_gen_loss,
            "young_gen_loss": total_young_gen_loss,
            "old_disc_loss": old_disc_loss,
            "young_disc_loss": young_disc_loss
        }
```

```python
with strategy.scope():
    def discriminator_loss(real, generated):
        real_loss = tf.keras.losses.BinaryCrossentropy(from_logits=True, reduction=tf.keras.losses.Reduction.NONE)(tf.ones_like(real), real)

        generated_loss = tf.keras.losses.BinaryCrossentropy(from_logits=True, reduction=tf.keras.losses.Reduction.NONE)(tf.zeros_like(generated), generated)

        total_disc_loss = real_loss + generated_loss

        return total_disc_loss * 0.5
```

```python
with strategy.scope():
    def generator_loss(generated):
        return tf.keras.losses.BinaryCrossentropy(from_logits=True, reduction=tf.keras.losses.Reduction.NONE)(tf.ones_like(generated), generated)
```

```python
with strategy.scope():
    def calc_cycle_loss(real_image, cycled_image, LAMBDA):
        loss1 = tf.reduce_mean(tf.abs(real_image - cycled_image))

        return LAMBDA * loss1
```

```python
with strategy.scope():
    def identity_loss(real_image, same_image, LAMBDA):
        loss = tf.reduce_mean(tf.abs(real_image - same_image))
        return LAMBDA * 0.5 * loss
```

```python
with strategy.scope():
    old_generator_optimizer = tf.keras.optimizers.Adam(2e-4, beta_1=0.5)
    young_generator_optimizer = tf.keras.optimizers.Adam(2e-4, beta_1=0.5)

    old_discriminator_optimizer = tf.keras.optimizers.Adam(2e-4, beta_1=0.5)
    young_discriminator_optimizer = tf.keras.optimizers.Adam(2e-4, beta_1=0.5)
```

```python
with strategy.scope():
    cycle_gan_model = CycleGan(
        old_generator, young_generator, old_discriminator, young_discriminator
    )

    cycle_gan_model.compile(
        o_gen_optimizer = old_generator_optimizer,
        y_gen_optimizer = young_generator_optimizer,
        o_disc_optimizer = old_discriminator_optimizer,
        y_disc_optimizer = young_discriminator_optimizer,
        gen_loss_fn = generator_loss,
        disc_loss_fn = discriminator_loss,
        cycle_loss_fn = calc_cycle_loss,
        identity_loss_fn = identity_loss
    )
```

```python
cycle_gan_model.fit(
    tf.data.Dataset.zip((old_ds, young_ds)),
    epochs=25
)
```
